```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
# Check data structure
df.info()
```

```python
# Features and target
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
target = 'Survived'
```

```python
# One-hot encoding
ohe = OneHotEncoder()
encoded_features = ohe.fit_transform(df[features]).toarray()  # This will throw an error since features have not been processed correctly (e.g., missing values)
encoded_features = pd.DataFrame(encoded_features)
```

```python
# Imputation: Apply different imputation techniques
imputer_mean = SimpleImputer(strategy='mean')
df['Age'] = imputer_mean.fit_transform(df[['Age']]).ravel()  # Imputation for 'Age'
```

```python
imputer_median = SimpleImputer(strategy='median')
df['Fare'] = imputer_median.fit_transform(df[['Fare']]).ravel()  # Imputation for 'Fare'
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(encoded_features, df[target], test_size=0.2, random_state=42)
```
